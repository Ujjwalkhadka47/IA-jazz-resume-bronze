```python
"""
2023-01-01
"""

from pyspark.sql import SparkSession
from pyspark.sql.functions import regexp_replace

spark = SparkSession.builder \
    .appName("Read Parquet File") \
    .getOrCreate()

parquet_file_path = "/Users/user/jazz-resume-bronze-dev/resume_text_ml_data_2023-01-01.parquet"

df = spark.read.parquet(parquet_file_path)

df.show()



```

```python

df = df.withColumn("resume_file_path", regexp_replace("resume_file_path", "output", "s3://fuse-ia-jazz-resume-bronze-dev"))


df.show(truncate=False)

```
